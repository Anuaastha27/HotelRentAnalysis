```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

1.Price Analysis by Neighborhood, Room Type, and Region

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
# Grouping by neighborhood_group and room_type, then calculating the average price
price_by_neighborhood_room = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().unstack()

# Plotting the average price by neighborhood and room type
price_by_neighborhood_room.plot(kind='bar', figsize=(10, 6))
plt.title('Average Price by Neighborhood and Room Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

2. Host Analysis: Multiple Listings

```python
# Convert host_id data into numpy array
host_ids = df['host_id'].values

# Calculate the number of listings for each unique host
unique_hosts, host_listing_counts = np.unique(host_ids, return_counts=True)

# Plotting using Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(host_listing_counts, bins=50)
plt.title('Distribution of Number of Listings per Host')
plt.xlabel('Number of Listings')
plt.ylabel('Frequency')
plt.show()
```

3. Review Patterns: Relationship Between Reviews and Price

```python
# Scatter plot between price and number of reviews
df.plot(kind='scatter', x='number_of_reviews', y='price', alpha=0.3, figsize=(8, 6))
plt.title('Relationship Between Number of Reviews and Price')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

4. Availability Analysis: Average Availability by Region and Room Type

```python
# Grouping by neighborhood_group and room_type, then calculating the average availability
availability_by_neighborhood_room = df.groupby(['neighbourhood_group', 'room_type'])['availability_365'].mean().unstack()

# Plotting the average availability by neighborhood and room type
availability_by_neighborhood_room.plot(kind='bar', figsize=(10, 6))
plt.title('Average Availability by Neighborhood and Room Type')
plt.ylabel('Average Availability (Days)')
plt.xticks(rotation=45)
plt.show()
```

5. Geospatial Analysis: Distribution of Listings

```python
# Scatter plot of listings on latitude and longitude
df.plot(kind='scatter', x='longitude', y='latitude', alpha=0.5, figsize=(10, 5),
        c='price', cmap='viridis', colorbar=True)
plt.title('Geospatial Distribution of Listings by Price')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
